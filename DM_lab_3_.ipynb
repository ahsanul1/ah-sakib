```python
import pandas as pd
ds = pd.read_csv ("/content/drive/MyDrive/Data Mining Lab/iris.csv")
```

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = ds ['species'].value_counts()

name = data.index
count = data.values

plt.title('Scores')
plt.xlabel('Names')
plt.ylabel('Total Count')

plt.bar(name,count)
```

```python
colors = {'setosa':'r','versicolor':'m','virginica':'b'}          #Dictionary of color
fig, ax = plt.subplots()

plt.title('Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

for i in range(len(ds ['sepal_width'])):
  ax.scatter( ds ['sepal_length'][i], ds ['sepal_width'][i],color=colors[ds ['species'][i]])
```

```python
slices = [4, 6, 5]

activity = ['setosa','versicolor','virginica']   

cols = ["g", "b", "r"]

plt.pie(slices,
        labels=activity,
        startangle=45,
        shadow=True,
        explode=(0.2,0.2,0.2),
        autopct="%1.1f%%")
plt.title("Scatter Plot")
plt.show()
```
